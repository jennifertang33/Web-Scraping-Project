```python
df['date']=pd.to_datetime(df['date'], dayfirst=True)   #converting ['date'] to datetime
df
```

```python
df['month_year'] = df['date'].dt.to_period('M')       #creating new column with only month and year
```

```python
df['month_year'].unique()
```

```python
df['month_year']=df['month_year'].map(str)      #['month_year'] is in a type called 'period'
                                                #change it back to string
```

```python
period = []                               #find out all unique month_year
for a in df['month_year']:
    period.append(a)
period1 = sorted(set(period))
type(period1)
```

```python
type(df['title'])
```

```python
for j in df['title']:
    j=j.lower()
```

```python
df['title']
```

```python
list_final=[]

for a in period1:

    list1 = []
    for i in df[df['month_year']==a]['title']:
        list1.append(i.replace('"', '').replace('(', '').replace(')', '').replace('/', ' ')
        .replace('\xa0', ' ').lower().split(' '))

    list2 = []
    for j in list1:
        for k in j:
            list2.append(k)

    list_final.append(list2)
```

```python
list_final
```

```python
list_final_1=[]

for b in list_final:
    
    list3 = Counter(b).most_common(20)
    list_final_1.append(list3)
```

```python
list_final_1
```

```python
dict1 = pd.DataFrame(list_final_1)
```

```python
dict2 = dict1.T
dict2.columns = period1
dict2.iloc[:, 26:38]
```
